# Demo RAG + generación de imágenes (GoT)

Este notebook usa `got_assistant.py` con los modelos de Hugging Face:
- `Qwen/Qwen3-30B-A3B-Thinking-2507-FP8`
- `stabilityai/stable-diffusion-3.5-large`

```python
from pathlib import Path
from got_assistant import GoTAssistant

HF_TOKEN = ""  # Pon tu token aquí o usa variable de entorno
EPUB_PATH = Path("book.epub")
QUESTION = "¿Cómo se describe Invernalia al inicio de la historia?"
```

```python
assistant = GoTAssistant(hf_token=HF_TOKEN)
assistant.load_book(EPUB_PATH)
result = assistant.ask(QUESTION, top_k=5)
result['answer']
```

```python
img_path = assistant.generate_image(
    positive_prompt=result['image_prompt']['positive_prompt'],
    negative_prompt=result['image_prompt']['negative_prompt'],
    output_path=Path('outputs/notebook_scene.png')
)
img_path
```
